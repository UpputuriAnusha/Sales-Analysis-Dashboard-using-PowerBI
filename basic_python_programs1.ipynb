```python
#finding maximum number between two numbers
a=int(input("enter first number"))
b=int(input("enter second number"))
if (a>b):
  print("greatest number is:",a)
else:
  print("greatest number is:",b)
```

```python
#finding maximum number between three numbers
a=int(input("enter first number"))
b=int(input("enter second number"))
c=int(input("enter third number"))
if (a>b and a>c):
  print("greatest number is:",a)
elif(b>a and b>c):
  print("greatest number is:",b)
else:
  print("greatest number is:",c)
```

```python
#finding positive,negative and zero
a=int(input("enter a number"))
if(a>0):
  print("it is a positive number",a)
elif(a<0):
  print("it is a negative number",a)
else:
  print("it is zero",a)
```

```python
#finding evenor odd
a=int(input("enter a number"))
if(a%2==0):
  print("it is an even number",a)
else:
  print("it is an odd number",a)
```

```python
#finding leap year or not
a=int(input("enter a year"))
if(a%400==0):
  print("it is a leap year",a)
else:
  print("it is a non leap year",a)
```

```python
#swap two variables using temporary variable
a=int(input("enter the value of first number(a):"))
b=int(input("enter the value of second number(b):"))
print(f"original values:a={a},b={b}")
temp=a
a=b
b=temp
print(f"swapped values:a={a},b={b}")
```

```python
#swap two variables without using temp variable
a=5
b=10
a,b=b,a
print("after swapping")
print("a=",a)
print("b=",b)
```

```python
#check prime number or not
n=int(input("enter a number"))
count=0
for i in range(1,n+1):
  if(n%i==0):
    count=count+1
if(count==2):
  print(n,"is prime")
else:
    print("is not prime")
```

```python
#check prime numbers between in interval
min=0
max=100
for i in range(min,max+1):
  count=0
  for j in range(1,i+1):
   if(i%j==0):
    count=count+1
  if(count==2):
    print(i,end=' ')

```
